```python
# Importamos esta libreria que nos permite reemplazar caracteres
import nltk
import re

nltk.download('punkt')
nltk.download('stopwords')

# Importamos la función que nos permite Stemmizar de nltk y definimos el stemmer
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

# Traemos nuevamente las stopwords
stopwords = nltk.corpus.stopwords.words('english')
#stopword.remove('no')
#stopword.remove('not')
```

```python
# Recorremos todos los registros y le vamos aplicando la Normalizacion y luega el Stemming a cada uno
registro_list=[]
for registro in data['overview']:
  # Vamos a reemplazar los caracteres que no sean letras por espacios
  registro=re.sub("[^a-zA-Z]"," ",str(registro))
    # Pasamos todo a minúsculas
  registro=registro.lower()
    # Tokenizamos para separar las palabras del titular
  registro=nltk.word_tokenize(registro)
    # Eliminamos las palabras de menos de 3 letras
  registro = [palabra for palabra in registro if len(palabra)>3]
    # Sacamos las Stopwords
  registro = [palabra for palabra in registro if not palabra in stopwords]
    
    ## Hasta acá Normalizamos, ahora a stemmizar
    
    # Aplicamos la funcion para buscar la raiz de las palabras
  registro=[stemmer.stem(palabra) for palabra in registro]
    # Por ultimo volvemos a unir el titular
  registro=" ".join(registro)
    
    # Vamos armando una lista con todos los titulares
  registro_list.append(registro)
data['overview_stem'] = registro_list
```

```python
# Recorremos todos los registros y le vamos aplicando la Normalizacion y luega el Stemming a cada uno
registro_list=[]
for registro in data['tagline']:
  # Vamos a reemplazar los caracteres que no sean letras por espacios
  registro=re.sub("[^a-zA-Z]"," ",str(registro))
    # Pasamos todo a minúsculas
  registro=registro.lower()
    # Tokenizamos para separar las palabras del titular
  registro=nltk.word_tokenize(registro)
    # Eliminamos las palabras de menos de 3 letras
  registro = [palabra for palabra in registro if len(palabra)>3]
    # Sacamos las Stopwords
  registro = [palabra for palabra in registro if not palabra in stopwords]
    
    ## Hasta acá Normalizamos, ahora a stemmizar
    
    # Aplicamos la funcion para buscar la raiz de las palabras
  registro=[stemmer.stem(palabra) for palabra in registro]
    # Por ultimo volvemos a unir el titular
  registro=" ".join(registro)
    
    # Vamos armando una lista con todos los titulares
  registro_list.append(registro)
data['tagline_stem'] = registro_list
```

```python
#Creamos una funcion para unir los nombres de cada elemento de la lista (eliminar espacio)
# ej: Tom Hans= TomHans
def collapse(L):
    L1 = []
    L=ast.literal_eval(L)
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
for columna in ['genres','cast','crew']:
  data[columna] = data[columna].apply(collapse)

# las columnas con string las pasamos a lista
data['overview_stem'] = data['overview_stem'].apply(lambda x: x.split())
data['tagline_stem'] = data['tagline_stem'].apply(lambda x: x.split())
data['title_'] = data['title'].apply(lambda x: x.split())
```

```python
# Concatenamos todas las columnas con etiquetas en una sola
data['tags'] =data['title_'] + data['genres']+ data['cast']+ data['crew'] + data['tagline_stem'] + data['overview_stem']
```

```python
#volvemos a pasar a string
data['tags'] = data['tags'].apply(lambda x: ' '.join(x))
```

```python
# nos quedamos con las columnas necesarias para el modelo 
data=data[['id','title','tags']]
data=data.reset_index(drop=True)
```

Sistema de recomendacion

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(reduced_data)
```

```python
def recommend(movie):
  movie= movie.strip()
  movie_index = data[data['title'] == movie].index[0] #index ) para obtener id
  distances = similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
  peliculas={}
  pelicula=[]
  for i in movie_list:
    pelicula.append(data.iloc[i[0]]['title'])
  peliculas['lista recomendada']=pelicula
  return peliculas
```

```python
recommend('Girls Trip')
```
